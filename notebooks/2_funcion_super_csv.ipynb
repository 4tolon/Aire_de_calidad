### Funciona!!! ahora vamos con el resto de archivos para juntar todos los meses hasta tener 6 years.

```python
anos = ['2016', '2017', '2018', '2019', '2020', '2021',]
```

```python
subpath = ['../data/1_am_ca/', '../data/2_cm_ca/']
```

```python
data_day_list = []
ensayo ='ensayo_8'
for i in range(70):
    a1 = pd.read_csv(f'../data/{ensayo}/{contenido[i]}', index_col=0)
    a1.index = pd.to_datetime(a1.index)
    a1 = a1.asfreq('H')
    c1 = pd.read_csv(f'../data/{ensayo}/{contenido[i+70]}', index_col=0)
    c1.index = pd.to_datetime(c1.index)
    c1 = c1.asfreq('H')
    merge=pd.merge(a1,c1, how='inner', left_index=True, right_index=True)
    data_day_list.append(merge)
    
final_df = pd.concat(data_day_list)
final_df.to_csv(f'../data/ensayo_8_.csv')
```
